```python
 # Pandas
import pandas as pd

# SQL Alchemy
from sqlalchemy import create_engine

# Define sql database name
database = "electricity_db"
```

```python
# Create path to password
from config import password
```

```python
 # Create Engine
engine = create_engine(f"postgresql://postgres:" + password + "@localhost/" + database)
conn = engine.connect()
```

```python
# Query All Records in the the continent table
continent_df = pd.read_sql("SELECT * FROM continent", conn)

# Query All Records in the the continent table
gdp_df = pd.read_sql("SELECT * FROM gdp", conn)

# Query All Records in the the continent table
population_df = pd.read_sql("SELECT * FROM population", conn)

# Query All Records in the the continent table
energyconsumption_df = pd.read_sql("SELECT * FROM energyconsumption", conn)
```

```python
energyconsumption_df.head(10)
```

```python
energyconsumption_final_df = pd.merge(energyconsumption_df,gdp_df,how="inner",
                                      left_on=['country_code','month','year'], right_on = ['country_code','month','year'])
```

```python
energyconsumption_final_df = energyconsumption_final_df.drop(columns={'country_y'})
energyconsumption_final_df = energyconsumption_final_df.rename(columns={'country_x':'country'})
energyconsumption_final_df
```

```python
energyconsumption_final_df["date"] = pd.to_datetime(energyconsumption_final_df['year'].astype(str) + energyconsumption_final_df['month'].astype(str)+
                                       "01" ,format='%Y%m%d')
```

```python
energyconsumption_final_df = pd.merge(energyconsumption_final_df,population_df,how="inner",
                                      left_on=['country_code','month','year'], right_on = ['country_code','month','year'])
```

```python
energyconsumption_final_df.columns
```

```python
energyconsumption_final_df = energyconsumption_final_df.drop(columns={'year','month','id_x','id_y','country_y','id'})
energyconsumption_final_df = energyconsumption_final_df.rename(columns={'country_x':'country'})
energyconsumption_final_df = energyconsumption_final_df.fillna(0)
energyconsumption_final_df
```

```python
energyconsumption_final_df.to_csv('FinalModelCSV/energyconsumptiongdppopulation.csv', index=False)
```
